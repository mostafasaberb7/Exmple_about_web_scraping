### task1

```python
import requests
from bs4 import BeautifulSoup

request = requests.get("https://www.exchangerates.org.uk/Dollars-to-Egyptian-Pounds-currency-conversion-page.html")

soup = BeautifulSoup(request.text, 'html.parser')

rate = soup.find('span', attrs={'id':'shd2b;'})
rate = float(rate.get_text())

amount_usd = float(input('Enter amount in USD'))
print(f'{amount_usd} amount in EGP = {amount_usd*rate}')
```

### task3

```python
import requests
from bs4 import BeautifulSoup
import csv

response = requests.get("https://books.toscrape.com/catalogue/category/books_1/index.html")
soup = BeautifulSoup(response.text, "html.parser") 

books1 = soup.find_all("article", attrs={"class":"product_pod"})
ratings = {'One':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5}

with open("books.csv","w",newline="",encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Name","Price","Rating"])
 
    for article in books1:
            name = article.find_all("a")[1].get('title')
            price = float(article.find('p', attrs={"class":"price_color"}).get_text()[2:])
            rating = article.find('p', attrs={"class":'star-rating'}).get('class')[1]
            rating = ratings[rating]
            print([name,price,rating])
            writer.writerow([name,price,rating])
    page = 1
    running = True

    while running:
        response = requests.get(f"https://books.toscrape.com/catalogue/category/books_1/page-{page}.html")
        soup = BeautifulSoup(response.text, "html.parser") 
        books = soup.find_all("article", attrs={"class":"product_pod"})

        if len(books) <= 0:
            running = False
            break
        ratings = {'One':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5}
        for article in books:
            name = article.find_all("a")[1].get('title')
            price = float(article.find('p', attrs={"class":"price_color"}).get_text()[2:])
            rating = article.find('p', attrs={"class":'star-rating'}).get('class')[1]
            rating = ratings[rating]
            print(name,price,rating)
            writer.writerow([name,price,rating])
        page +=1
```

### task4

```python
import requests
from bs4 import BeautifulSoup
import csv
response = requests.get("https://www.scrapethissite.com/pages/simple/")
soup = BeautifulSoup(response.text, "html.parser") 

div0 = soup.find_all("div" , attrs={"class":"container"})[1]
div1 = div0.find_all("div",attrs={"class":"col-md-4 country"})

with open("countries.csv" ,"w", newline="",encoding="utf-8")as file:
    writer = csv.writer(file)
    writer.writerow(["Name","Capital","Population","Area-km2"])
    for i in div1:
        name = i.find("h3").get_text().strip()
        capital = i.find("span" ,attrs={"class":"country-capital"}).get_text().strip()
        population = i.find("span" ,attrs={"class":"country-population"}).get_text().strip()
        area = i.find("span" ,attrs={"class":"country-area"}).get_text().strip()
        row = [name,capital,population,area]
        writer.writerow(row) 
        print(row)
```
